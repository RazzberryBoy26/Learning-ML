```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
z0 = -2 + 0j
z1 = 0 + 2j
z2 = 2 + 0j
z_values = [z0, z1, z2]
i = 0;
while(i < 497):
    zn = z_values[len(z_values) - 3] + (z_values[len(z_values) - 2] - z_values[len(z_values) - 3])/3
    z_values.append(zn)
    i += 1
df = pd.DataFrame(z_values, columns = ["Z"])
df["Re"] = df["Z"].apply(lambda x: x.real)
df["Im"] = df["Z"].apply(lambda x: x.imag)
```

```python
plt.plot(df["Re"], df["Im"], label = "Trajectory")
plt.gca().set_aspect('equal')
plt.scatter(df["Re"], df["Im"], alpha = 0.3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.legend()
plt.show()
```
