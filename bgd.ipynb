```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
import sklearn as sk
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
X_raw = data.data
Y = data.target.reshape(-1, 1)
scaler  = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)
```

```python
print(data.feature_names)
```

```python
def grad(X, Y, Q):
    H = X @ Q
    return (X.T @ (H - Y))/X.shape[0]
m = X_scaled.shape[0]
ones = np.ones((m, 1))
X = np.hstack([ones, X_scaled])
Q = np.zeros((X.shape[1], 1))
a = 1e-3
iters = 20000
i = 0
Qz = [0 + 0j]
while(i < iters):
    Q = Q - a*grad(X, Y, Q)
    zn = Q[0, 0] + Q[1, 0]*1j
    Qz.append(zn)
    if(zn - Qz[len(Qz) - 2] < 0):
        break;
    i += 1
print(i)
```

```python
df = pd.DataFrame(Qz, columns = ["Z"])
df["Re"] = df["Z"].apply(lambda x: x.real)
df["Im"] = df["Z"].apply(lambda x: x.imag)
H = X @ Q
plt.scatter(Y, H, alpha=0.3)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Predicted vs Actual California Housing Prices")
plt.grid(True)
plt.show()
plt.plot(df["Re"], df["Im"], label="Trajectory", linewidth=1)
plt.xlabel("Q0")
plt.ylabel("Q1")
plt.title("Batch Gradient Descent Parametric convergence")
plt.xlim([0, 2.2])
plt.ylim([0, 2.2])
plt.gca().set_aspect('equal')
plt.grid(True)
plt.legend()
plt.show()
print(Q[0, 0])
print(Q[1, 0])
```
